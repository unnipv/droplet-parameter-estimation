```python
# Load image, grayscale, Threshold, return cropped 
def crop_defocus(imgdata, df):
    image = cv2.imread(imgdata['path'])
    original = image.copy()
    gray = cv2.blur(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY),(7,7))
    thresh = cv2.threshold(gray, 138, 255, cv2.THRESH_BINARY_INV)[1]
    prefix = "E:\Academics\DDP\Data\Droplet50_100_150\Dataset\\"
    # Find contours, obtain bounding box, extract and save ROI
    cnts = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cnts = cnts[0] if len(cnts) == 2 else cnts[1]

    #Creating bounding boxes for the detected countours, creating 256 x 256 crops of the ROIs
    ROI_number = 0
    for cnt in cnts:
        x,y,w,h = cv2.boundingRect(cnt)
        # Rejecting ROIs that are not roughly square shaped to avoid low-quality training data
        if w/h > 0.9 and w/h < 1.1 and w*h > 2500:
            x_center = int(0.5*(2*x + w))
            y_center = int(0.5*(2*y + h))
            x_left = max(int(x_center - 128), 0)
            y_left = max(int(y_center - 128), 0)
            x_right = min(int(x_center + 128), image.shape[1])
            y_right = min(int(y_center + 128), image.shape[0])
            ROI = original[y_left:y_right, x_left:x_right]
            
            # Rejecting ROIs that are not of 256 x 256 size
            if ROI.shape[0] != 256 or ROI.shape[1] != 256:
                continue
            
            #Saving the file and saving file name and labels to a dataframe
            crop_file_path = prefix + imgdata['name'].split('.tif')[0] + '_{}.png'.format(ROI_number)
            crop_name = imgdata['name'].split('.tif')[0] + '_{}.png'.format(ROI_number)
            cv2.imwrite(crop_file_path, ROI)
            
            a_series = pd.Series([crop_name, imgdata['dia'], imgdata['pos']], index = df.columns)
            df = df.append(a_series, ignore_index=True)
            ROI_number += 1
    return df
```

```python
# Load image, grayscale, Threshold, return the ROI pasted on a 512 x 512 bg
def crop_defocus_paste(imgdata, df, memo, prefix):
    bg = cv2.imread(r"E:\Academics\DDP\Data\Defocused\bg\bg.png")
    image = cv2.imread(imgdata['path'])
    original = image.copy()
    gray = cv2.blur(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY),(7,7))
    
# Different threshold conditions for different diameters
    if imgdata['dia'] == 400:
        thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]
    else:
        thresh = cv2.threshold(gray, 138, 255, cv2.THRESH_BINARY_INV)[1]

# Find contours, obtain bounding box, extract and save ROI
    cnts = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cnts = cnts[0] if len(cnts) == 2 else cnts[1]

# Creating bounding boxes for the detected countours, creating 256 x 256 crops of the ROIs
    ROI_number = 0
    for cnt in cnts:
        x,y,w,h = cv2.boundingRect(cnt)
        
# Rejecting ROIs that are not roughly square shaped to avoid low-quality training data
        if w/h > 0.9 and w/h < 1.1 and w*h > 2500:
            if (memo[imgdata['dia']][imgdata['pos'] if imgdata['pos'] >= 0 else (300 - imgdata['pos'])]) >= 100:
                break
            else:
                memo[imgdata['dia']][imgdata['pos'] if imgdata['pos'] >= 0 else (300 - imgdata['pos'])] += 1
            x_center = int(0.5*(2*x + w))
            y_center = int(0.5*(2*y + h))
            x_left = max(int(x_center - w*0.7), 0)
            y_left = max(int(y_center - h*0.7), 0)
            x_right = min(int(x_center + w*0.7), image.shape[1])
            y_right = min(int(y_center + h*0.7), image.shape[0])
            ROI = original[y_left:y_right, x_left:x_right]
            
            # Pasting the ROI on 512 x 512 background
            ROI = cv2.cvtColor(ROI, cv2.COLOR_BGR2RGB)
            back_im = Image.fromarray(bg).copy()
            origin = ((512 - ROI.shape[1])//2, (512 - ROI.shape[0])//2)
            ROI_pil = Image.fromarray(ROI)
            back_im.paste(ROI_pil, origin)
            
            #Saving the file and saving file name and labels to a dataframe
            crop_file_path = prefix + imgdata['name'].split('.tif')[0] + '_{}.png'.format(ROI_number)
            crop_name = imgdata['name'].split('.tif')[0] + '_{}.png'.format(ROI_number)
            back_im.save(crop_file_path)
            
            a_series = pd.Series([crop_name, imgdata['dia'], imgdata['pos']], index = df.columns)
            df = df.append(a_series, ignore_index=True)
            ROI_number += 1
    return df, memo
```
