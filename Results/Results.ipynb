```python
listOfCSV = [infAll, infNo100, infNo200, De100, De5115Dia, De5115Pos, DeAllDia, DeAllPos, DeNo200Dia, DeNo200Pos,
             DeNo100Dia, DeNo100Pos]

stats = [
    {
        'name' : 'infAll'
    },
    {
        'name' : 'infNo100'
    },
    {
        'name' : 'infNo200'
    },
    {
        'name' : 'De100'
    },
    {
        'name' : 'De5115Dia'
    },
    {
        'name' : 'De5115Pos'
    },
    {
        'name' : 'DeAllDia'
    },
    {
        'name' : 'DeAllPos'
    },
    {
        'name' : 'DeNo200Dia'
    },
    {
        'name' : 'DeNo200Pos'
    },
    {
        'name' : 'DeNo100Dia'
    },
    {
        'name' : 'DeNo100Pos'
    }
]
```

```python
def rmse(y_true, y_pred):
    MSE = np.square(np.subtract(y_true, y_pred)).mean() 
    RMSE = math.sqrt(MSE)
    return RMSE

def mape(y_true, y_pred):
    diff = np.abs((y_true - y_pred)/y_true)
    diff = diff[diff < 1000]
    mape = np.mean(diff)*100
    return mape

def rmspe(y_true, y_pred):
    diff = np.square(((y_true - y_pred) / y_true))
    diff = diff[diff < 1000]
    return np.sqrt(np.nanmean(diff)*100)
```

```python
for i, df in enumerate(listOfCSV):
    stats[i]['RMSE'] = rmse(df['Label'], df['Predicted'])
    stats[i]['MAPE'] = mape(df['Label'], df['Predicted'])
    stats[i]['RMSPE'] = rmspe(df['Label'], df['Predicted'])
```
